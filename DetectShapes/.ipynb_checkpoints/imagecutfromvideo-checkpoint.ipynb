```python
import sys
import argparse
import matplotlib.pyplot as plt

import cv2
print(cv2.__version__)

def extractImages(pathIn, pathOut):
    count = 0
    vidcap = cv2.VideoCapture(pathIn)
    success,image = vidcap.read()
    success = True
    while success:
        vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*1000))    # added this line 
        success,image = vidcap.read()
        print ('Read a new frame: ', success)
        cv2.imwrite( pathOut + r"\frame%d.jpg" % count, image)     # save frame as JPEG file
        count = count + 1

if __name__=="__main__":
    a = argparse.ArgumentParser()
    a.add_argument(r"C:\Users\davidra\Desktop\NINJA_RAFAEL\Ninja2-master\DetectShapes\pics\20220627_104515.mp4", help="path to video")
    a.add_argument(r"C:\Users\davidra\Desktop\NINJA_RAFAEL\Ninja2-master\DetectShapes\pics_from_mp4", help="path to images")
    args = a.parse_args()
    print(args)
    extractImages(args.pathIn, args.pathOut)
```
